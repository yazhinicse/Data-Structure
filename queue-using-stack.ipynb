```python
class Queue:
    
    def __init__(self):
        self.s1 = []  #push stack
        self.s2 = []  # pop stack
        self.size = 0
 
    def enqueue(self, data):
        self.size += 1
        # Move all elements from push stack to pop stack
        while len(self.s1) != 0:  # checking whether push stack is empty or not.
            self.s2.append(self.s1[-1]) # take out the last element in the list and push it to the pop stack
            self.s1.pop() # deleting the last element from the push stack
 
        self.s1.append(data)  # push item into push stack
 
        while len(self.s2) != 0: # push everything back to push stack from pop stack
            self.s1.append(self.s2[-1]) # take out last element from the pop stack and push it to the push stack
            self.s2.pop() # delete the last element (top) from the pop stack
        
    def dequeue(self):
        self.size -= 1
        if len(self.s1) == 0: # checking whether the queue is empty or not
            print("Queue is empty. You cannot delete anything from the queue")        
        else:
            x = self.s1[-1] 
            self.s1.pop()
            return x
    
    def display_queue(self):
        print(self.s1)
    
    def get_size(self):
        return len(self.s1)
```

```python
if __name__ == '__main__':
    
    queue = Queue() # Queue object
    
    while True:
        option = input("Enter your option:\n 1. Enqueue\n 2. Dequeue\n 3. Quit\n")
        option = int(option.lower())
        
        if option == 1:     
            val = input("Enter the value: ")
            queue.enqueue(val)
            queue.display_queue()
            print("Current size of the queue is ", queue.get_size())
            print("----------------------------\n")
        elif option == 2:
            queue.dequeue()
            queue.display_queue()
            print("Current size of the queue is ", queue.get_size())
            print("----------------------------\n")
        elif option == 3:
            break
```
