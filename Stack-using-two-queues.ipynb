```python
class Stack:
    def __init__(self): # constructor for the stack
        self.q1 = Queue() # initilalizing the first queue
        self.q2 = Queue() # initilalizing the second queue
        self.s_size = 0 # to store the size of the stack
        
    def is_empty(self): # to check whether a stack is empty or not
        return self.q2.is_empty()
    
    def push(self, data): # to insert an element into the stack
        self.s_size += 1 # increase the size of the stack when a new data is inserted
        self.q1.enqueue(data) # insert a new element into the queue
        
        while not self.q2.is_empty(): # to transfer all the elements from q2 to q1
            x = self.q2.dequeue()
            self.q1.enqueue(x)
        
        # to swap the contents in q1 and q2
        self.q = self.q1 
        self.q1 = self.q2 
        self.q2 = self.q
        
    def pop(self): # to delete an element from the stack
        if self.is_empty():
            print("Stack is empty!!! You cannot delete anything from the stack!!!")
        else:
            self.s_size -= 1  # reduce the size of the stack when an element is removed
            return self.q2.dequeue() # remove an element from the queue
    
    def display_stack(self): # to print the contents in the queue
        if self.q2.is_empty(): # cannot display if the stack is empty
            return
        return self.q2.display_queue() # get the elements from the queue and display it
    
    def get_size(self): # to get the size of the stack
        return self.s_size
```

```python
class Queue:
    def __init__(self): # a constructor
        self.items = []  # to store the elements in a queue
        self.q_size = 0 # to compute the size of the queue
        
    def is_empty(self): # checking whether a queue is empty or not
        return self.items == []
    
    def enqueue(self, data): # to insert a new data to the queue
        self.q_size += 1 # increasing the size of a queue when a new data is inserted
        #self.q_size = self.q_size + 1
        self.items.append(data) # inserting new data to the queue list 
        
    def dequeue(self): # to remove an element from the queue
        self.q_size -= 1 # reducing the size of the queue when an element is removed from the queue
        return self.items.pop(0) # delete the items from the front. pop() uses the index 0 to delete the element from the front
    
    def get_size(self): # to get the size of the queue 
        return self.q_size
    
    def display_queue(self): # to print the elements in the queue
        print(self.items[0:self.q_size]) # return all the elements in the queue
```

```python
if __name__ == '__main__':
    
    stack = Stack() # initialzing a stack object
      
    while True:
        
        option = input("Enter your option (PUSH/POP/Quit):") # read input from the keyboard
        option = option.lower() # converts all the characters into lower case. It is a Python function
        
        if option == "push" :
            val = input("Enter the value: ")
            stack.push(val)      
            stack.display_stack()
            print("Current size of the stack is ", stack.get_size())
            print("----------------------------\n")
    
        elif option == "pop":
            stack.pop()
            stack.display_stack()
            print("Current size of the stack is ", stack.get_size())
            print("----------------------------\n")
            
        elif option == "quit":
            break
```
