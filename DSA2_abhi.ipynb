Problem 1:
Alice is rearranging her library. She takes the innermost shelf and reverses the order of books. If you want to automate this process, which data structure would you prefer to do this and why? 


Problem 2:
There are two persons A and B. Both are maintaining two book shelves. A takes the first book (the book he kept for the first time in the shelf) from the shelf and reads it. B prefers to take the last book and reads. After reading a book, both A and B keep the book in a separate shelf. Write a Python program to implement this process. Identify the data structures and define the operations accordingly.

Problem 1 Answer:
To automate the process, I would prefer stack as the data structure, because what we can do is: we can keep popping the elements untill we reach the innermost shelf and then reverse the order and after reversing the order we can push back the popped ones.




Problem 2

Since A takes the first book he kept for the first time, its best to use the data structure with FIFO implementation. Therefore Queue is the suitable DS

Now coming to B, he prefers to take the last book to read. So its nothing but but taking the book present at the top. So Stack is the suitable data structure here


```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class AShelfQueue:
    
    def __init__(self):
        self.front=self.rear=None
        #initially they point to none
        
    def isEmpty(self):
        if self.front == None:
            return True
        else:
            return False
        
    def Enqueue(self,data):
        new_node = Node(data)
        #while inserting we consider rear
        if self.rear == None:
            self.front = new_node
            self.rear = new_node
            return
        #when newnode comes, i need to change the pointers
        self.rear.next=new_node
        self.rear=new_node
        
    
    def Dequeue(self):
        if self.isEmpty():
            print("No books in A's Book Shelf")
        else:
            temp = self.front
            #i store the read book in temp
            #so that i can add to separate shelf later
            self.front = temp.next
            temp.next = None
            print("\n",temp.data,"The book read by A will be kept in separate shelf")
            return temp.data
        
            if(self.front == None):
                self.rear = None
                
    def printAShelf(self):
        node = self.front
        if self.isEmpty():
            print("There are no books in A's Shelf")
        else:
            print("Books in A 's shelf are", end="\n")
            while(node != None):
                print(node.data, end = "\n")
                node = node.next
            return
```

```python
class BStackShelf:
    def __init__(self):
        self.start = None
        
    def isEmpty(self):
        if self.start == None:
            return True
        else:
            return False
        
    def push(self,data):
        if self.start == None:
            self.start = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.start
            self.start = new_node
            
    def pop(self):
        if self.isEmpty():
            print("Book shelf of B is empty")
        else:
            #here again i am storing the read book because
            #i want to put it into new shelf
            temp = self.start
            self.start = self.start.next
            temp.next = None
            print("\n",temp.data, "Keep the read book in separate shelf")
            return temp.data
        
    def printBShelf(self):
        node = self.start
        if self.isEmpty():
            print("No books in B's shelf")
        else:
            print("books in B's helf are \n")
            while(node!=None):
                  print(node.data, end = "\n")
                  node = node.next
            return
```

```python
if __name__ == '__main__':
    
    new_shelf = list()
    #as no operations is mentioned on new shelf
    #will use the list instead of creating a new stack or queue
    Shelf_of_A = AShelfQueue()
    Shelf_of_B = BStackShelf()
    
    while True:
        
        choice = input("\n Enter your choice\n 1:Insert book in Shelf of A \n 2: Insert Book in Shelf of B \n 3: A reads \n 4: B reads \n 5: Quit \n ")
        
        
        if choice == "1":
            book = input("Enter the book to insert in to A's shelf: ")
            Shelf_of_A.Enqueue(book)
            Shelf_of_A.printAShelf()
        
        elif choice == "2":
            book = input("Enter the book to insert in B's Shelf: ")
            Shelf_of_B.push(book)
            Shelf_of_B.printBShelf()
            
        elif choice == "3":
            print("A reads the book")
            item = Shelf_of_A.Dequeue()
            new_shelf.append(item)
            print("Books in new shelf",new_shelf)
            
        elif choice == "4":
            print("B reads the book")
            item = Shelf_of_B.pop()
            new_shelf.append(item)
            print("Booksin new shelf", new_shelf)
            
        elif choice == "5":
            #pass
            break
        
            
            
            
    
```
