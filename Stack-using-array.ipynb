```python
class Stack:
    
    def __init__(self):
        self.stack = list()
        self.MAXSIZE = 5
        self.TOP = 0
    
    def isEmpty(self):
        if self.TOP == 0:
            return True
        else:
            return False
            
    def push(self, data):
        if self.TOP == self.MAXSIZE:
            print("Stack Full! Can not add any more elements\n")
        else:
            self.stack.append(data)
            self.TOP += 1
            
    
    def pop(self):
        if self.isEmpty():
            print("Stack Empty!\n")
        else:
            item = self.stack[self.TOP-1]
            self.TOP -= 1
            self.stack.pop()
            print(item, "has been removed from the stack\n")
        
    def printStack(self):
        for i in range(self.TOP):
            print(self.stack[i], end =" ")
        print()
        
    def size(self):
        print(len(self.stack))
        
    def topStack(self):
        print(self.stack[self.TOP-1])
```

```python
if __name__ == '__main__':
    s = Stack()
    s.push(4)
    s.printStack()
    s.push(6)
    s.printStack()
    s.push(7)
    s.printStack()
    s.pop()
    s.printStack()
    s.push(9)
    s.printStack()
    s.pop()
    s.printStack()
    s.pop()
    s.printStack()
    s.pop()
    s.printStack()
    s.pop()
    s.printStack()
    s.push(8)
    s.printStack()
    s.push(17)
    s.printStack()
    s.push(65)
    s.printStack()
    s.push(13)
    s.printStack()
    s.push(90)
    s.printStack()
    s.push(89)
    s.size()
    s.topStack()
```
